```python
import pandas as pd
import zipfile
import numpy as np
import matplotlib.pyplot as plt
```

```python
ABMD = pd.read_csv("C:/Users/sklen/Desktop/ABMD.csv")
BKNG = pd.read_csv("C:/Users/sklen/Desktop/BKNG.csv")
CPB = pd.read_csv("C:/Users/sklen/Desktop/CPB.csv")
HBAN = pd.read_csv("C:/Users/sklen/Desktop/HBAN.csv")
XOM = pd.read_csv("C:/Users/sklen/Desktop/XOM.csv")
```

```python
df1 = ABMD.merge(BKNG, how = "outer", on =["Date"])
df1
```

```python
df2 = df1.merge(CPB, how = "outer", on =["Date"])
df2
```

```python
df3 = df2.merge(HBAN, how = "outer", on =["Date"], suffixes=["_H","_N"])
df3
```

```python
df4 = df3.merge(XOM, how = "outer", on =["Date"])
df4
```

```python
# Logaritmic loss
df4['Log loss_x'] = np.log(df4['Adj Close_x'] / df4['Open_x'])
highest_gain = df4.loc[df4['Log loss_x'].idxmax()]
highest_loss = df4.loc[df4['Log loss_x'].idxmin()]
df4['Log loss_y'] = np.log(df4['Adj Close_y'] / df4['Open_y'])
highest_gain = df4.loc[df4['Log loss_y'].idxmax()]
highest_loss = df4.loc[df4['Log loss_y'].idxmin()]
df4['Log loss_H'] = np.log(df4['Adj Close_H'] / df4['Open_H'])
highest_gain = df4.loc[df4['Log loss_H'].idxmax()]
highest_loss = df4.loc[df4['Log loss_H'].idxmin()]
df4['Log loss_N'] = np.log(df4['Adj Close_N'] / df4['Open_N'])
highest_gain = df4.loc[df4['Log loss_N'].idxmax()]
highest_loss = df4.loc[df4['Log loss_N'].idxmin()]
df4['Log loss'] = np.log(df4['Adj Close'] / df4['Open'])
highest_gain = df4.loc[df4['Log loss'].idxmax()]
highest_loss = df4.loc[df4['Log loss'].idxmin()]

```

```python
df4['log_ret_x'] = np.log(df4['Adj Close_x'] ) - np.log(df4['Adj Close_x'].shift(1))
df4['log_ret_y'] = np.log(df4['Adj Close_y'] ) - np.log(df4['Adj Close_y'].shift(1))
df4['log_ret_H'] = np.log(df4['Adj Close_H'] ) - np.log(df4['Adj Close_H'].shift(1))
df4['log_ret_N'] = np.log(df4['Adj Close_N'] ) - np.log(df4['Adj Close_N'].shift(1))
df4['log_ret'] = np.log(df4['Adj Close'] ) - np.log(df4['Adj Close'].shift(1))
# x for ABMD, y for BKNG H for CPB, N for HBAN, log_ret for XOM
```

```python
highest_log_ret_x = df4['log_ret_x'].max()
highest_log_ret_y = df4['log_ret_y'].max()
highest_log_ret_H = df4['log_ret_H'].max()
highest_log_ret_N = df4['log_ret_N'].max()
highest_log_ret = df4['log_ret'].max()
low_log_ret_x = df4['log_ret_x'].min()
low_log_ret_y = df4['log_ret_y'].min()
low_log_ret_H = df4['log_ret_H'].min()
low_log_ret_N = df4['log_ret_N'].min()
low_log_ret = df4['log_ret'].min()
avg_log_ret_x = df4['log_ret_x'].mean()
avg_log_ret_y = df4['log_ret_y'].mean()
avg_log_ret_H = df4['log_ret_H'].mean()
avg_log_ret_N = df4['log_ret_N'].mean()
med_log_ret = df4['log_ret'].mean()
med_log_ret_x = df4['log_ret_x'].median()
med_log_ret_y = df4['log_ret_y'].median()
med_log_ret_H = df4['log_ret_H'].median()
med_log_ret_N = df4['log_ret_N'].median()
med_log_ret = df4['log_ret'].median()
```
